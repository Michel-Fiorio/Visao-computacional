```python
#Importando bibliotecas
import numpy as np
import cv2
from google.colab.patches import cv2_imshow     ## a função cv2_imshow do open cv não funciona naturalmente no colab, portanto tem que importar dos pacotes do google
from matplotlib import pyplot as plt
```

```python
# Carregando Imagem colorida
img = cv2.imread('/content/drive/MyDrive/gaivota.jpg')
cv2_imshow(img)
```

```python
# Obtendo informações da imagem
print ('Dimensoes da Imagem (linhas, colunas e bandas):', img.shape)
print ('Tamanho da Imagem em bytes (linhas x colunas x bandas):', img.size)
print ('Valores das bandas BGR na linha 100 e coluna 100:', img[99,99])
print ('Valor da banda azul na linha 100 e coluna 100:', img[99,99,0])
```

```python
# Carregando a imagem no formato escala de cinza
img2 = cv2.imread('/content/drive/MyDrive/gaivota.jpg', cv2.IMREAD_GRAYSCALE)
cv2_imshow(img2)
```

```python
# Obtendo informações da imagem monocromática
print ('Dimensoes da Imagem (linhas, colunas e bandas):', img2.shape)
print ('Tamanho da Imagem em bytes (linhas x colunas x bandas):', img2.size)
print ('Valores das bandas BGR na linha 100 e coluna 100:', img2[99,99])
```

```python
#Visualizando imagens grandes com matplot

plt.imshow(img)
plt.xticks([]), plt.yticks([]) # esconder os valors de Ox e Oy #esconder o eixo
plt.show()

#O OpenCV lê a imagem no sistema BGR, o Matplotlib trabalha no sistema RGB. Então, temos que converter a imagem no momento de visualizar dentro do Matplolib.
```

```python
#Visualizando imagens grandes com Matplot
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#Visualizar as duas imagens ao mesmo tempo
titulos = ['RGB', 'Monocromática']
imagens = [img, img2]

#0,1 em uma lista
for i in list(range(2)): # i varia de 0 a 1
    plt.subplot (1,2,i+1), plt.imshow(cv2.cvtColor(imagens[i], cv2.COLOR_BGR2RGB))
    plt.title(titulos[i])
    plt.xticks([]), plt.yticks([])
plt.show()
```
